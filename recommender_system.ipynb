```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
```

```python
products = pd.read_csv("products.csv")
orders = pd.read_csv("order_products__train.csv")
```

```python
# merge the two datasets
df = pd.merge(orders, products, on='product_id')
df.head(5)
```

```python
trans = df.groupby('order_id')['product_name'].apply(list)
```

```python
# since the original dataset encoding exceeds RAM
# use a sample
sampled_df = trans.sample(n=100000, random_state=42)
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()

# encode
te_ary = te.fit(sampled_df).transform(sampled_df)
df2 = pd.DataFrame(te_ary, columns=te.columns_)
df2.head(5)
```

```python
# use apriori to get the frequent items
frequent_itemsets = apriori(df2, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

# set the rules as confidence = 0.01
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.01)
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

```python
my_rules = rules[(rules['confidence'] > 0.2) & (rules['support'] > 0.015)]
my_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```
